## 1 matplotlib概述
### 1.1 介绍
专门用于开发2D图表（包括3D图表），使用起来及其简单，以渐进、交互方式实现数据可视化。
可视化是在整个数据挖掘的关键辅助工具，可以清晰的理解数据，从而调整我们的分析方法。
能够将数据进行可视化，更直观的呈现；使数据更加客观、更具说服力。


```python
# 实现一个简单的matplotlib画图
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.plot([1,0,9],[4,5,6])
plt.show()
```

### 1.2 三层结构
容器层包含（1）画板层Canvas（2）画布层 Figure（3）绘图区/坐标系Axes

Canvas是位于最底层的系统层，在绘图的过程中充当画板的角色，即放置画布(Figure)的工具。

Figure是 Canvas上方的第一层，也是需要用户来操作的应用层的第一层，在绘图的过程中充当画布的角色。

Axes是应用层的第二层，在绘图的过程中相当于画布上的坐标系/绘图区的角色，可以通过使用plt.subplots()来分割画布来得到若干个坐标系/绘图区，返回的是绘画区的列表，通过调用列表中每个绘画区在对应的位置画图；注意：设置绘画区的编号应该从编号1开始的。坐标系Axis这个很容易和Axes混淆，实际上它是位于每个Axes上，可以通过其设置显示图形所在坐标轴的大小、刻度、刻度标签。

辅助显示层
在这一层，用户可以设置添加标题、网格、图例等对图形的描述信息，它不会影响图形的显示效果。

图像层
在该层，正式绘图（直方图、折线图、散点图、饼状图）
注意：Matplotlib的三层结构是按层堆叠的。

### 2.3 创建多个绘图区


```python
# 创建画布
# plt.figure(figsize=(20,8),dpi=90)
figure,axes = plt.subplots(nrows=1,ncols=2,figsize=(20,8),dpi=80)


# 准备数据
y_beijin = y_shanhai = np.random.uniform(12,14,60)
y_shanhai = np.random.uniform(15,18,60)
x = np.arange(1,61,1)


# 绘制图像
axes[0].plot(x,y_shanhai,color='r',label="shanhai")
axes[1].plot(x,y_beijin,label="beijin")


# 修改x,y刻度
# 构造y轴刻度
y_ticks= range(30)
# 构造x轴刻度标签
x_ticks_label = ['11点{}分'.format(i) for i in x]


# 显示图例
axes[0].legend()
axes[1].legend()


# 修改y,y轴坐标的刻度显示
axes[0].set_xticks(x[::5])
axes[0].set_xticklabels(x_ticks_label[::5])
axes[0].set_yticks(y_ticks[::5])
axes[1].set_xticks(x[::5])
axes[1].set_xticklabels(x_ticks_label[::5])
axes[1].set_yticks(y_ticks[::5])

# 添加网格，以方便阅读
axes[0].grid(True,linestyle="--",alpha=0.5)
axes[1].grid(True,linestyle="-.",alpha=0.5)

# 添加描述信息
axes[0].set_xlabel("time")
axes[0].set_ylabel("tep")
axes[0].set_title("shanghaitep-time温度变化")

axes[1].set_xlabel("time")
axes[1].set_ylabel("tep")
axes[1].set_title("beijingtep-time温度变化")

plt.show()
```

### 柱状图
统计，更加直观的呈现

绘制柱状图
matplotlib.pyplot.bar(x,width,align='center',**kwargs)
x;有几个类别
y：在下例中，代表票房数量
width:柱状图的宽度
align：每个柱状图的位置对齐方式
color：选择柱状图的颜色


```python
# 准备数据
movie_name = ['雷神','正义联盟','东方快车','寻梦环游记','全球风暴','降魔传','追捕','七一']
tickets = [78345,22234,56789,34567,34568,35689,76543,69436]
# 创建画布
plt.figure(figsize=(20,8),dpi=80)

# 绘制柱状图
x_ticks = range(len(movie_name))
plt.bar(x_ticks,tickets,color=['b','g','r','b','y','k','m'])

# 修改x刻度
plt.xticks(x_ticks,movie_name)


# 添加网格显示
plt.grid(linestyle='--',alpha=0.5)

# 显示图像
plt.show()
```

两个柱状图联合

```python
# 准备数据
movie_name = ['a','b','c','d']
first_day = [20,30,50,20]
first_weekend = [70,40,60,30]
# 创建画布
plt.figure(figsize=(20,8),dpi=80)

# 绘制图像
x = range(len(movie_name))
plt.bar(range(4),first_day,width=0.2,color='r',label='first_day')
plt.bar([i+0.2 for i in x],first_weekend,width=0.2,color='b',label='first_weekend')

#显示图例
plt.legend()

# 修改刻度
# 修改x刻度
plt.xticks([i+0.1 for i in x],movie_name)
# 增加网格
plt.grid(linestyle='--',alpha=0.5)

# 显示图像
plt.show()
```

### 直方图
柱状图和直方图的区别。
直方图：用于反应一组连续数据的分布状况。
直方图形状类似柱状图却有着与柱状图完全不同的含义。直方图牵涉统计学得概念，首先要对数据进行分组，然后统计每个分组内数据元得数量。在坐标系中，横轴标出每个组的端点，纵轴表示频数，每个矩形的高代表对应的频数。

直方图绘制：
plt.hist(x,bins=None,normed=None,**kwargs)
x;原数据
bins:将数据分成几组
bins = max(x) - min(x) / 组距


概念：
组数：在统计数据时，我们把数据按照不同的范围分成几个组，分成的组的个数称为组数；
组距：每一组两个端点的差；


```python
# 250部电影的时长分布
# 准备数据
time = np.random.randint(80,150,250)
time
```

```python
# 创建画布
plt.figure(figsize=(20,8),dpi=80)

# 绘制直方图
distance = 2
group_num = int((max(time) - min(time)) / distance)

plt.hist(time,bins=group_num)


# 修改x轴刻度
plt.xticks(range(min(time),max(time),distance))


plt.show()
```

### 饼图
不同类别的占比情况
饼图api
注意显示的百分比位数
plt.pie(x,labels=,autopact,colors)
x:数量，自动算百分比
labels:每部分的名称
autopact:占比显示指定%1.2f% %
colors:每部分颜色

```python
#准备数据
movie_name =['a','b','c','d','e']
place_count = [3456,6543,7201,3478,4589]

# 创建画布
plt.figure(figsize=(20,8),dpi=80)

# 绘制饼图
plt.pie(place_count,labels=movie_name,autopct="%1.2f%%",colors=['r','y','m','k','b'])

# 为了让显示的饼图保持图形，需要添加axis保证长宽一样
plt.axis('equal')

# 显示图例
plt.legend()


# # 显示图像
plt.show()
```
