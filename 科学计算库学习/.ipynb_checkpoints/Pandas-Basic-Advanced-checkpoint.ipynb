### 处理缺失值

处理缺失值的方法：
1、删除含有缺失值的样本；df.dropna(inplace=False)默认按行来删除，不会修改原数据，需要接受返回值
2、替换法；使用均值填补；df.fillna(value,inplace=False)
在处理完成后，需要对填补后的数据进行判断是否还有缺失；
注意：不是缺失值nan，有默认标记的，需要将默认标记先替换为nan。

```python
# 方法1：删除含有缺失值的样本
new_movie = movie.dropna() # 默认inplace参数为False，返回新的DataFrame
```

```python
pd.isnull(new_movie).any()   # 新返回的DataFrame已经不含有缺失值
```

```python
pd.isnull(movie).any() # 原DataFrame未变，仍含有缺失值
```

```python
# 方法2：替换法
# 含有缺失值的字段 Revenue (Millions) Metasore
# 使用均值填补 Revenue (Millions) 字段的缺失值
movie['Revenue (Millions)'].fillna(value=movie['Revenue (Millions)'].mean(),inplace=True)
```

```python
# 使用均值填补Metascore字段缺失值
movie['Metascore'].fillna(value=movie['Metascore'].mean(),inplace=True)
```

```python
# 判断填补后的movie是否存在缺失值
print(pd.isnull(movie).any())
```

## 数据离散化

数据离散化概念：连续属性的离散化就是将连续属性的值域上，将值域划分为若干个离散的区间，最后用不同的符号或整数值代表落在每个子区间中的属性值；

对数据离散化的原因：连续属性离散化的目的是为了简化数据结构，数据离散化技术可以用来减少给定连续属性值的个数。离散化方法经常作为数据挖掘的工具

实现对数据离散化的步骤：
1、对数据分组：
①自动分组pd.qcut(data,bins)
②自定义分组pd.cut(data,bins)
③对数据进行分组将数据分组：一般会value_counts搭配使用，统计每组的个数。series.value_counts()统计分组次数

2、对分好组的数据求哑变量
pandas.get_dummies(data.prefix=None)
prefix分组名字

### 按照索引拼接

pd.merge(left,right,how='inner',on = None,left_on=None,right_on=None,left_index=False,right_index=False,sort=Ture,suffixes=('_x','_y'),copy=Ture,indicator=False,validate=None)

可以指定按照两组数据的共同键值对合并或者左右各自；
on = None按照哪一个字段进行索引拼接
left,right两组数据；
left_on=None,right_on=None指定左右键；

```python
left = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                     'key2':['K0','K1','K0','K1'],
                     'A':['A0','A1','A2','A3'],
                     'B':['B0','B1','B2','B3']})
```

```python
left
```

```python
right = pd.DataFrame({'key1':['K0','K1','K1','K2'],
                      'key2':['K0','K0','K0','K0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
right
```

```python
# 内连接 数据取两个DataFrame的交集
pd.merge(left,right, how='inner',on=['key1', 'key2'])
```

```python
# 左连接 左侧DataFrame取全部数据
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

```python
# 右连接 右侧DataFrame取全部数据
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```python
# 外连接 数据取两个DataFrame的并集
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```python
# 进行分组，对color分组，对price1进行聚合
# 用dataframe的方法进行分组
data = col.groupby(by='color')
```

```python
data
```

```python
data['price1'].max()
```

```python
# 用series的方法进行分组
data= col['price1'].groupby(by=col['color'])
```

```python
data
```

```python
data.max()
```

```python
# 对color、object分组，对price1进行聚合
data = col.groupby(by=['color', 'object'])
```

```python
data['price1'].max()
```

星巴克店铺数据

```python
# 获取数据
starbucks = pd.read_csv('directory.csv')
```

```python
starbucks
```

```python
# 进行分组聚合
# 按照国家分组，求助每个国家的星巴克零售店数量
count = starbucks.groupby(['Country'])
```

```python
count.count()
```

```python
starbucks.groupby(['Country']).count()['Brand'].plot(kind='bar')
```

```python
# 排序，降序排列,取前十个
starbucks.groupby(['Country']).count()['Brand'].sort_values(ascending=False)[:10].plot(kind='bar')
```

```python
# 加入城市字段，一起索引
starbucks.groupby(['Country','State/Province']).count()
```
